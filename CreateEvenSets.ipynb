```
import numpy as np
import glob
import gc
```

```
#get the file directory for train data
FileDirX= '/content/drive/MyDrive/CS 766 Project/Project Coding and Data Files/Data Batches/Stratified Random Sampling Datasets/Xtrain/*.npy'
FileDirY= '/content/drive/MyDrive/CS 766 Project/Project Coding and Data Files/Data Batches/Stratified Random Sampling Datasets/Ytrain/*.npy'

Files_ListX = glob.glob(FileDirX)
Files_ListY = glob.glob(FileDirY)

Files_ListX.sort()
Files_ListY.sort()

for k in range(0, len(Files_ListY)):
  print(Files_ListX[k], '\n')
  print(Files_ListY[k], '\n')
```

```
def GetData(XFile, YFile):

  print(XFile, '\n')
  print(YFile, '\n')
  Xtrain = []
  Ytrain = []

  data = np.load(XFile, allow_pickle = True) # load the train data
  Y = np.load(YFile, allow_pickle=True) # load the labels

  #extract even amount of data from each class (we choose 62 because of the limiting class-4)
  dataNew0 = data[0:62]
  dataNew1 = data[2293: (2293+62)]
  dataNew2 = data[2293+216:2293+216+62]
  dataNew3 = data[2293+216+470:2293+216+470+62]
  dataNew4 = data[2293+216+470+77:]

  Xtrain = Xtrain + [x for x in dataNew0]
  Xtrain = Xtrain + [x for x in dataNew1]
  Xtrain = Xtrain + [x for x in dataNew2]
  Xtrain = Xtrain + [x for x in dataNew3]
  Xtrain = Xtrain + [x for x in dataNew4]

  #load the Y labels
  YNew0 = Y[0:62]
  YNew1 = Y[2293: (2293+62)]
  YNew2 = Y[2293+216:2293+216+62]
  YNew3 = Y[2293+216+470:2293+216+470+62]
  YNew4 = Y[2293+216+470+77:]

  Ytrain = Ytrain + [x for x in YNew0]
  Ytrain = Ytrain + [x for x in YNew1]
  Ytrain = Ytrain + [x for x in YNew2]
  Ytrain = Ytrain + [x for x in YNew3]
  Ytrain = Ytrain + [x for x in YNew4]

  #Xtrain = Xtrain + [dataNew0] + [dataNew1] + [dataNew2] + [dataNew3] + [dataNew4]
  #Ytrain = Ytrain + [YNew0] + [YNew1] + [YNew2] + [YNew3] + [YNew4] 

  data = None
  del data
  gc.collect()

  np.save('/content/drive/MyDrive/CS 766 Project/Project Coding and Data Files/Data Batches/Even Class Distribution Datasets/X' + XFile[125:-4] + '_even', Xtrain)
  np.save('/content/drive/MyDrive/CS 766 Project/Project Coding and Data Files/Data Batches/Even Class Distribution Datasets/Y' + YFile[125:-4] + '_even', Ytrain)
```

```
for k in range(0,len(Files_ListX)):
  GetData(Files_ListX[k], Files_ListY[k])
```

```
## test data batches 

#get the file directory for train data
FileDirX= '/content/drive/MyDrive/CS 766 Project/Project Coding and Data Files/Data Batches/Stratified Random Sampling Datasets/Xtest/*.npy'
FileDirY= '/content/drive/MyDrive/CS 766 Project/Project Coding and Data Files/Data Batches/Stratified Random Sampling Datasets/Ytest/*.npy'
Files_ListX = glob.glob(FileDirX)
Files_ListY = glob.glob(FileDirY)

Files_ListX.sort()
Files_ListY.sort()

for k in range(0, len(Files_ListY)):
  print(Files_ListX[k], '\n')
  print(Files_ListY[k], '\n')
```

```

#create different parsing function since last set is slightly larger (data doesnt divide perfectly into batches)
def GetDataLastTestBatch(XFile, YFile):

  print(XFile, '\n')
  print(YFile, '\n')
  Xtrain = []
  Ytrain = []

  data = np.load(XFile, allow_pickle = True) # load the train data
  Y = np.load(YFile, allow_pickle=True) # load the labels

  #extract even amount of data from each class (we choose 62 because of the limiting class-4)
  dataNew0 = data[0:88]
  dataNew1 = data[2872:(2872+88)]
  dataNew2 = data[2872+278:2872+278+88]
  dataNew3 = data[2872+278+588:2872+278+588+88]
  dataNew4 = data[2872+278+588+102:]

  Xtrain = Xtrain + [x for x in dataNew0]
  Xtrain = Xtrain + [x for x in dataNew1]
  Xtrain = Xtrain + [x for x in dataNew2]
  Xtrain = Xtrain + [x for x in dataNew3]
  Xtrain = Xtrain + [x for x in dataNew4]

  #load the Y labels
  YNew0 = Y[0:88]
  YNew1 = Y[2872:(2872+88)]
  YNew2 = Y[2872+278:2872+278+88]
  YNew3 = Y[2872+278+588:2872+278+588+88]
  YNew4 = Y[2872+278+588+102:]

  Ytrain = Ytrain + [x for x in YNew0]
  Ytrain = Ytrain + [x for x in YNew1]
  Ytrain = Ytrain + [x for x in YNew2]
  Ytrain = Ytrain + [x for x in YNew3]
  Ytrain = Ytrain + [x for x in YNew4]

  #Xtrain = Xtrain + [dataNew0] + [dataNew1] + [dataNew2] + [dataNew3] + [dataNew4]
  #Ytrain = Ytrain + [YNew0] + [YNew1] + [YNew2] + [YNew3] + [YNew4] 

  data = None
  del data
  gc.collect()

  np.save('/content/drive/MyDrive/CS 766 Project/Project Coding and Data Files/Data Batches/Even Class Distribution Datasets/X' + XFile[125:-4] + '_even', Xtrain)
  np.save('/content/drive/MyDrive/CS 766 Project/Project Coding and Data Files/Data Batches/Even Class Distribution Datasets/Y' + YFile[125:-4] + '_even', Ytrain)
```

```
#do first test batch
GetData(Files_ListX[0], Files_ListY[0])
GetDataLastTestBatch(Files_ListX[1], Files_ListY[1]) # do the last test batch separately, which has slightly more samples
```
